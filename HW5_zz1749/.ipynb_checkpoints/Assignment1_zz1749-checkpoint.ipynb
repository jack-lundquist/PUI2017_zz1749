```python
import pandas as pd
import numpy as np
import pylab as pl
%pylab inline
import seaborn as sns
import scipy
```

```python
np.random.seed(999)
```

```python
# Generate N samples from a distribution
# sample size 
n = 500
# number of samples
N = 100
mymean = 20
# I'm going to choose Chi-square distribution
samples = {}
```

```python
mystd = np.sqrt(2*mymean)
```

```python
# values are the samples and their keys are their means
for i in range(N):
    sample = np.random.chisquare(mymean,n)
    samples[np.mean(sample)] = sample
```

```python
means = samples.keys()
```

```python
# plot the distribution of the first sample
sns.distplot(samples[samples.keys()[0]],label='probability distribution');
pl.xlabel('samples')
pl.ylabel('distribution')
pl.title('chisquare')
pl.legend()
pl.show()
```

## Figure 3. the distribution of my z statistics

## H0: the z statistics follow a N(0,1) distribution

```python
# Fit the z-statistics with a gaussian model and assess the goodness of fit 
# with a simple test

print ('z-scores on normal', scipy.stats.kstest(z,'norm',args=[0,1]))
print ('z-scores on normal', scipy.stats.anderson(z,'norm'))
```

## From the ks test, we could see that the p-value is way bigger than 0.05 so we cannot reject the null hypothesis that the z statistics follow a N(0,1) distribution
## From the ad test, we could see that the ad statistics is 0.3169 which is smaller than the corresponding critical value: 0.759 so we also cannot reject the null hypothesis that the z statistics follow a N(0,1) distribution

```python
# now we are going to set our arguments as mean and std
def chi2(mymean):
    np.random.seed(999)
    n = 500
    N = 100
    samples = {}
    mystd = np.sqrt(2*mymean)
    for i in range(N):
        sample = np.random.chisquare(mymean,n)
        samples[np.mean(sample)] = sample
    means = samples.keys()
    z = []
    [z.append((i-mymean)*np.sqrt(n)/mystd) for i in samples.keys()];
    hist, bin_edges = np.histogram(z, normed=True)
    bincenters = bin_edges[:-1] + 0.5*(bin_edges[1] - bin_edges[0]) 
    exp = scipy.stats.norm.pdf(bincenters)
    chi2 = scipy.stats.chisquare(hist,exp)[0]
    return chi2
```

```python
chi2(23)
```

```python
result = scipy.optimize.minimize(chi2,x0 =mymean)
```
